+ **Outlier in Activity type**

```python
def IQR_outlier(s):
    q1, q3 = s.quantile(.25), s.quantile(.75)
    iqr = q3 - q1
    low, up = q1 - 3*iqr, q3 + 3*iqr
    outlier_num = sum((s<low) | (s>up))
    return outlier_num
```

```python
# Created daily Activity Outlier table.
dailyActivityOutlier = dailyActivity.groupby('Id', as_index=True).agg(IQR_outlier)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(dailyActivityOutlier.sum(), dailyActivityOutlier.sum().index)
plt.title('Daily activity attribut Outlier', size=14)
```

Describe: All of main features have outliers.

+ **Outlier in sleep type**

```python
# Created sleep day Outlier table.
sleepDayOutlier = sleepDay.groupby('Id', as_index=True)['TotalMinutesAsleep','TotalTimeInBed'].agg(IQR_outlier)
```

```python
sleepDayOutlier.sum()
plt.figure(figsize=(6,3))
sns.barplot(sleepDayOutlier.sum(), sleepDayOutlier.sum().index)
```
